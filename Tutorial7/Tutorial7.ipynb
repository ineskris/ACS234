```python
def func( t, y ): 
    return (t**2 + 1) 
      
# Function for euler formula 
def euler( t0, y, h, t ): 
    temp = -0
  
    # Iterating till the point at which we 
    # need approximation 
    while t0 < t: 
        temp = y 
        y = y + h * func(t0, y) 
        t0 = t0 + h 
    return(y)

# Initial Values 
t0 = 1
y0 = 4
h = 0.5
  
# Value of x at which we need approximation 
t = 2
y = euler(t0, y0, h, t) 
print("Approximate solution at x = ", t, " is ", "%.6f"% y)
```

```python
ts = np.linspace(2, 7, 100)
ys = [euler(t0, y0, h, i) for i in ts]
```

```python
def analytic_sol(t):
    return(t**3/3 + t +8/3)

ytruth = [analytic_sol(i) for i in ts]
```

```python
plt.plot(ts, ys, label="Euler approximate h = %0.2f  " % h );
plt.plot(ts, ys, label="Euler approximate h = %0.2f  " % h );
plt.plot(ts, ytruth, label='Analytic solution')
plt.legend();
plt.title("Euler's Method");
```

```python
for count, h in enumerate([ 1,0.5, 0.2, 0.1, 0.001]):
    ys = [euler(t0, y0, h, i) for i in ts]
    ytruth = [analytic_sol(i) for i in ts]
    plt.xlabel('Time')
    plt.plot(ts, ys, label="Euler approximate h = %0.2f  " % h, color='red' );
    plt.plot(ts, ytruth, label='Analytic solution', color='black')
    plt.legend();
    plt.title("Euler's Method");
    plt.show()
```

```python
def func(t, y): 
    return (t**2 + 1) 
      
        
class Euler():
    def __init__(self, t0=t0, y0=y0, h=h, func=func):
        self.t0 = t0
        self.y0 = y0
        self.h = h
        self.func = func
  
    def euler(self, t): 
        temp = -0
        y = self.y0
        t0 = self.t0
        while t0 < t: 
            y = y + self.h * func(t0, y) 
            t0 = t0 + self.h 
        return(float(y))

```
